```python
#Model building
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
#create new column to identify high and low income
df['High/Low'] =np.where(df['Salary']=='>50K','High','Low')
```

```python
# drop unused column for the model
df = df.drop(columns = ['fnlwgt','Salary'])
```

```python
#display top 5 rows of the dataframe
df.head()
```

```python
x = pd.get_dummies(df,columns = ['Workclass','Education','Marital_status','Occupation','Relationship','Race','Sex','Native_country'])
```

```python
x = x.drop(columns = ['High/Low'])
```

```python
y = pd.get_dummies(df['High/Low'], drop_first = False)
```

```python
#Train, test split, sets aside 25% of the samples in the original dataset for testing.
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=1)
```

```python
k=40
knn = KNeighborsClassifier(n_neighbors=k,metric='euclidean')
knn.fit(X_train,y_train)
```

```python
#predict type of Iris based on all attributes in the dataset
y_pred = knn.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter
import pandas as pd
import matplotlib.ticker as ticker
%matplotlib inline
```

```python
acc = []
# Will take some time
from sklearn import metrics
for i in range(1,50):
    neigh = KNeighborsClassifier(n_neighbors = i).fit(X_train,y_train)
    yhat = neigh.predict(X_test)
    acc.append(metrics.accuracy_score(y_test, yhat))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,50),acc,color = 'blue',linestyle='dashed', 
         marker='o',markerfacecolor='red', markersize=10)
plt.title('accuracy vs. K Value')
plt.xlabel('K')
plt.ylabel('Accuracy')
print("Maximum accuracy:",max(acc),"at K =",acc.index(max(acc)))
```
